# Task
Create an interactive GUI in Google Colab using `ipywidgets` to manage a collection of question-answer pairs. The GUI should include input fields for questions and answers, and buttons for 'Frage speichern' (Save Question), 'Vorblättern' (Next), 'Zurückblättern' (Previous), and 'alles speichern' (Save All). The system should initialize by attempting to load previously saved question-answer pairs from a pickled file and save the current state to a pickled file upon clicking 'alles speichern'. Provide a summary of the GUI's functionalities and instructions on how to use it.

## Initialize GUI and Data Structure

### Subtask:
Set up the `ipywidgets` for question input, answer input, and the 'Save Question', 'Next', 'Previous', and 'Save All' buttons. Initialize an empty nested list `qa_pairs` to store the question-answer pairs and an `current_index` variable to track the currently displayed pair. Also, attempt to load any previously saved data from a pickled file to populate `qa_pairs` and set the initial `current_index`.


**Reasoning**:
The subtask requires initializing ipywidgets for input and buttons, an empty list for Q&A pairs, an index tracker, and loading existing data from a pickle file. All these steps can be consolidated into a single Python code block.



```python
from ipywidgets import widgets
from IPython.display import display
import pickle
import os

# 2. Create Textarea widgets for question and answer input
question_input = widgets.Textarea(
    value='',
    placeholder='Geben Sie hier Ihre Frage ein',
    description='Frage:',
    disabled=False,
    layout=widgets.Layout(width='auto', height='auto')
)

answer_input = widgets.Textarea(
    value='',
    placeholder='Geben Sie hier Ihre Antwort ein',
    description='Antwort:',
    disabled=False,
    layout=widgets.Layout(width='auto', height='auto')
)

# 3. Create Button widgets
save_question_button = widgets.Button(
    description='Frage speichern',
    button_style='success',
    tooltip='Speichert die aktuelle Frage und Antwort',
    layout=widgets.Layout(width='auto')
)

next_button = widgets.Button(
    description='Vorblättern',
    button_style='info',
    tooltip='Zeigt die nächste Frage an',
    layout=widgets.Layout(width='auto')
)

previous_button = widgets.Button(
    description='Zurückblättern',
    button_style='info',
    tooltip='Zeigt die vorherige Frage an',
    layout=widgets.Layout(width='auto')
)

save_all_button = widgets.Button(
    description='Alles speichern',
    button_style='warning',
    tooltip='Speichert alle Frage-Antwort-Paare in einer Datei',
    layout=widgets.Layout(width='auto')
)

new_question_button = widgets.Button(
    description='Neue Frage',
    button_style='success', # Changed to success as it's a primary action
    tooltip='Löscht den Inhalt der Eingabefelder und bereitet eine neue Frage vor',
    layout=widgets.Layout(width='auto')
)

# 4. Create an Output widget
output_area = widgets.Output()

# 5. Initialize an empty list named qa_pairs
qa_pairs = []

# 6. Initialize an integer variable current_index
current_index = 0

# 7. Use a try-except block to attempt loading qa_pairs from 'qa_data.pkl'
try:
    with open('qa_data.pkl', 'rb') as f:
        loaded_data = pickle.load(f)
        if isinstance(loaded_data, list):
            qa_pairs = loaded_data
            with output_area:
                print(f"'{len(qa_pairs)}' Frage-Antwort-Paare erfolgreich geladen.")
        else:
            with output_area:
                print("Warnung: Die geladenen Daten sind nicht im erwarteten Format (Liste).")

except FileNotFoundError:
    with output_area:
        print("Keine vorhandene 'qa_data.pkl' Datei gefunden. Startet mit leeren Daten.")
except Exception as e:
    with output_area:
        print(f"Fehler beim Laden der Daten: {e}")

# 8. If qa_data.pkl is loaded successfully and qa_pairs is not empty, update widgets
if qa_pairs:
    current_index = 0 # Ensure current_index is 0 for the first loaded pair
    question_input.value = qa_pairs[current_index][0]
    answer_input.value = qa_pairs[current_index][1]
    with output_area:
        print(f"Anzeige von Frage {current_index + 1} von {len(qa_pairs)}.")

print("GUI widgets and data structures initialized.")
```

## Implement 'Frage speichern' (Save Question) Functionality

### Subtask:
Create a function to be called when the 'Save Question' button is clicked. This function will retrieve the text from both the question and answer input fields. If the `current_index` corresponds to an existing entry, it will update that entry in `qa_pairs`. If it's a new entry (or `qa_pairs` is empty), it will append the new question-answer pair to `qa_pairs`. After saving, clear the input fields and advance the `current_index` to allow for a new entry.


**Reasoning**:
The subtask requires implementing the `on_save_question_button_clicked` function, which involves retrieving input, validating it, updating or appending to the `qa_pairs` list, adjusting `current_index`, clearing input fields, and attaching the function to the 'Save Question' button. All these steps can be consolidated into a single Python code block.



```python
def on_save_question_button_clicked(b):
    global current_index, qa_pairs
    with output_area:
        output_area.clear_output()
        question = question_input.value.strip()
        answer = answer_input.value.strip()

        if not question or not answer:
            print("Bitte geben Sie sowohl eine Frage als auch eine Antwort ein.")
            return

        if current_index < len(qa_pairs):
            # Update existing entry
            qa_pairs[current_index] = [question, answer]
            print(f"Frage {current_index + 1} aktualisiert. Gesamt: {len(qa_pairs)} Paare.")
            # After updating, ensure the updated values are displayed
            question_input.value = qa_pairs[current_index][0]
            answer_input.value = qa_pairs[current_index][1]
        else:
            # Add new entry
            qa_pairs.append([question, answer])
            current_index = len(qa_pairs) - 1 # Point to the newly added item
            print(f"Neues Frage-Antwort-Paar gespeichert. Gesamt: {len(qa_pairs)} Paare.")
            # After adding, display the newly added item
            question_input.value = qa_pairs[current_index][0]
            answer_input.value = qa_pairs[current_index][1]

        # Output current position feedback
        print(f"Anzeige von Frage {current_index + 1} von {len(qa_pairs)}.")

# Attach the function to the save_question_button's click event
save_question_button.on_click(on_save_question_button_clicked)

print("'on_save_question_button_clicked' function defined and attached.")
```

## Implement 'Vorblättern' (Next) Functionality

### Subtask:
Create a function for the 'Next' button. This function will increment the `current_index` if it's not already at the end of the `qa_pairs` list. It will then display the question and answer from `qa_pairs[current_index]` in the respective input fields.


**Reasoning**:
The subtask requires implementing the `on_next_button_clicked` function, which involves navigating through the `qa_pairs` list, updating the `current_index`, displaying the corresponding question and answer, and attaching the function to the 'Next' button. All these steps can be consolidated into a single Python code block.



```python
def on_next_button_clicked(b):
    global current_index, qa_pairs
    with output_area:
        output_area.clear_output()
        if not qa_pairs:
            print("Es gibt keine gespeicherten Fragen zum Anzeigen.")
            question_input.value = ''
            answer_input.value = ''
            current_index = 0
            return

        if current_index < len(qa_pairs) - 1:
            current_index += 1
            question_input.value = qa_pairs[current_index][0]
            answer_input.value = qa_pairs[current_index][1]
            print(f"Anzeige von Frage {current_index + 1} von {len(qa_pairs)}.")
        else:
            print("Sie sind bereits bei der letzten Frage.")

# Attach the function to the next_button's click event
next_button.on_click(on_next_button_clicked)

print("'on_next_button_clicked' function defined and attached.")
```

## Implement 'Zurückblättern' (Previous) Functionality

### Subtask:
Create a function for the 'Previous' button. This function will decrement the `current_index` if it's not already at the beginning of the `qa_pairs` list. It will then display the question and answer from `qa_pairs[current_index]` in the respective input fields.


**Reasoning**:
The subtask requires implementing the `on_previous_button_clicked` function, which involves navigating through the `qa_pairs` list backwards, updating the `current_index`, displaying the corresponding question and answer, and attaching the function to the 'Previous' button. All these steps can be consolidated into a single Python code block.



```python
def on_previous_button_clicked(b):
    global current_index, qa_pairs
    with output_area:
        output_area.clear_output()
        if not qa_pairs:
            print("Es gibt keine gespeicherten Fragen zum Anzeigen.")
            question_input.value = ''
            answer_input.value = ''
            current_index = 0
            return

        if current_index > 0:
            current_index -= 1
            question_input.value = qa_pairs[current_index][0]
            answer_input.value = qa_pairs[current_index][1]
            print(f"Anzeige von Frage {current_index + 1} von {len(qa_pairs)}.")
        else:
            print("Sie sind bereits bei der ersten Frage.")

# Attach the function to the previous_button's click event
previous_button.on_click(on_previous_button_clicked)

print("'on_previous_button_clicked' function defined and attached.")
```

## Implement 'alles speichern' (Save All) Functionality

### Subtask:
Create a function for the 'Save All' button. This function will serialize the `qa_pairs` list to a file named `qa_data.pkl` using the `pickle` module. It should also provide feedback to the user regarding the success or failure of the save operation.

**Reasoning**:
The subtask requires implementing the `on_save_all_button_clicked` function, which involves serializing the `qa_pairs` list to a pickle file, providing user feedback, and attaching the function to the 'Save All' button. All these steps can be consolidated into a single Python code block.



```python
def on_save_all_button_clicked(b):
    global qa_pairs
    with output_area:
        output_area.clear_output()
        try:
            with open('qa_data.pkl', 'wb') as f:
                pickle.dump(qa_pairs, f)
            print(f"{len(qa_pairs)} Frage-Antwort-Paare erfolgreich in 'qa_data.pkl' gespeichert.")
        except Exception as e:
            print(f"Fehler beim Speichern der Daten: {e}")

# Attach the function to the save_all_button's click event
save_all_button.on_click(on_save_all_button_clicked)

print("'on_save_all_button_clicked' function defined and attached.")
```

**Reasoning**:
All button functionalities have been implemented. The next step is to display the widgets to the user so they can interact with the GUI, which is a core part of the main task.
